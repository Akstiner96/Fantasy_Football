```python
import pandas as pd
import numpy as np
from statsmodels.tsa.arima_model import ARIMA
from pathlib import Path
%matplotlib inline
```

```python
filepath = Path('./data.csv')
data = pd.read_csv(filepath)
```

```python
data
```

```python
#Breaking data up by position so that I can drop useless columns for that position
def positions():
    qb_data = []
    rb_data = []
    wr_data = []
    te_data = []
    fb_data = []
    dst_data = []
    k_data = []
    
    
    for i, row in data.iterrows():
        if data.at[i, 'Pos'] == 'QB':
            qb_data.append(row)
        
        elif data.at[i, 'Pos'] == 'RB':
            rb_data.append(row)
            
        elif data.at[i, 'Pos'] == 'WR':
            wr_data.append(row)
        
        elif data.at[i, 'Pos'] == 'TE':
            te_data.append(row)
        
        elif data.at[i, 'Pos'] == 'FB':
            fb_data.append(row)
            
        elif data.at[i, 'Pos'] == 'DST':
            dst_data.append(row)
            
        else:
            k_data.append(row)
            
            
    qb_df = pd.DataFrame(qb_data)        
    rb_df = pd.DataFrame(rb_data)
    wr_df = pd.DataFrame(wr_data)
    te_df = pd.DataFrame(te_data)
    fb_df = pd.DataFrame(fb_data)
    dst_df = pd.DataFrame(dst_data)
    k_df = pd.DataFrame(k_data)
    
    return qb_df, rb_df, wr_df, te_df, fb_df, dst_df, k_df
            
```

```python
qb, rb, wr, te, fb, dst, k = positions()
```
